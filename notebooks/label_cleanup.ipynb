# Yorliq Tozalash (Label Cleanup)

Bu notebook `data/label.csv` faylidagi yorliqlarni tozalash va tekshirish uchun.

- Dublikat rasmlarni topish
- Noto'g'ri yorliqlarni aniqlash
- Mavjud bo'lmagan fayllarni filtrlash

```python
import os
import pandas as pd

# Label faylni yuklash
df = pd.read_csv('../data/label.csv')
print(f'Jami yozuvlar: {len(df)}')
df.head()
```

```python
# Sinf taqsimoti
print(df['choice'].value_counts())
print(f'\nDublikatlar: {df.duplicated().sum()}')
```

```python
# Mavjud bo'lmagan rasmlarni tekshirish
raw_dir = '../data/raw_images'
missing = []
for img in df['image']:
    if not os.path.exists(os.path.join(raw_dir, img)):
        missing.append(img)

print(f'Topilmagan rasmlar: {len(missing)}')
if missing:
    print(missing[:10])
```
